# Gradient Decent



# In  Predefined    function   finding   Gradient   Decent 

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt

#initializing variables
arr=[]
i = 0
x0= 4
x1= 0
learning= 0.1

while(i<=100): #if(i>100 then break)
    
    fx1= 2*x0

    x1= x0 - learning*fx1   # gradient Decent formula

    arr.append(x0)
    if(x0-x1 < 1e-5): #checking the vaules to break
        print('point - ',x1)
        break
    else:
        x0= x1
        i=i+1  #incrementing i value
        

if(i==100): 
    print('100 times over')
    print('point - ',x1)
    
X_values = np.arange(-5,5,.2)  

p = np.array([x**2 for x in X_values])

#ploting the curve
plt.xlabel("X")
plt.ylabel("f(x)")
plt.title("f(x) = x*2")
plt.plot(X_values,p)
plt.grid()


# ploting the point which lends to zero
for x in arr:
    plt.scatter(x,x**2)
```

# In  Dynamic function   finding   Gradient   Decent 

```python
#defining the dynamic function
fun = lambda x: x**3
df = lambda x: 3* x**2
```

```python
#GD function for any functions f(x)
def GD_Function(fun,df,ini,lr=0.01,tol=1e-5):
    i=0
    x1=0
    arr=[]


    while(i<100):

        x1= ini - lr*df(ini)  #formula for gd
        
        if((ini - x1) < tol):
            print('point - ', x1)
            break  # break when i==100
        else:
            ini = x1
            i=i+1
        arr.append(ini)

    #p = np.array([fun(x) for x in arr])
    X_values = np.arange(-5,5,.2)
    p = np.array([fun(x) for x in X_values])

    #ploting the function
    plt.plot(X_values,p)

    #ploting the curve
    plt.xlabel("X")
    plt.ylabel("f(x)")
    plt.title("f(x) = x^3")
    plt.grid()

    
    #ploting the values which leads to zero
    for x in arr:
        plt.scatter(x,fun(x))
    
    if(i==100):
        print('100 times over')
        print('point - ', x1)
        
```

```python
GD_Function(fun,df,5)
```
